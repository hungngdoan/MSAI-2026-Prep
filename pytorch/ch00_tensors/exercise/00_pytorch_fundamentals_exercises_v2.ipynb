### 2. Create a random tensor with shape `(5, 5)`.

````python
# Import torch
import torch

# Create random tensor
tensor55=torch.rand(5,5)
print(tensor55)
````

### 3. Perform a matrix multiplication on the tensor from 2 with another random tensor with shape `(1, 5)` (hint: you may have to transpose the second tensor).

````python
# Create another random tensor
tensor15 = torch.rand(1,5)

# Perform matrix multiplication
tensor51=tensor15.T
matmulRes = torch.matmul(tensor55, tensor51)
print (matmulRes)
print (matmulRes.shape)
````

### 4. Set the random seed to `42` and do 2 & 3 over again.

The resulting tensor should have shape `torch.Size([5, 1])`. Use `torch.manual_seed(42)` before creating both tensors so the results are reproducible.

````python
# Set manual seed

torch.manual_seed(42)
# Create two random tensors
tensor55=torch.rand(5,5)
tensorB=torch.rand(1,5)
matmulRes = torch.matmul(tensor55, tensorB.T)
print (matmulRes)
print (matmulRes.shape)

# Matrix multiply tensors
````

### 6. Create two random tensors of shape `(3, 4)` and send them both to the GPU (you'll need access to a GPU for this). Set `torch.manual_seed(99)` when creating the tensors (this doesn't have to be the GPU random seed).

The expected output structure should be something like:

```
Device: cuda
(tensor([[..., ..., ..., ...],
         [..., ..., ..., ...],
         [..., ..., ..., ...]], device='cuda:0'),
 tensor([[..., ..., ..., ...],
         [..., ..., ..., ...],
         [..., ..., ..., ...]], device='cuda:0'))
```

````python
# Set random seed
torch.manual_seed(99)

# Check for access to GPU
device = "cuda" if torch.cuda.is_available() else "cpu"

tensor_A = torch.rand(3,4).to(device)
tensor_B = torch.rand(3,4).to(device)
print(f"Device: {device}")
print(tensor_A)
print(tensor_B)

# Create two random tensors on GPU
````

### 7. Perform a matrix multiplication on the tensors you created in 6 (again, you may have to adjust the shapes of one of the tensors).

The resulting tensor should have shape `torch.Size([3, 3])`.

````python
# Perform matmul on tensor_A and tensor_B
output = tensor_A @ tensor_B.T
print(output)
print(output.shape)
````

### 8. Find the maximum and minimum values of the output of 7.

````python
# Find max
print(output.max())


# Find min
print(output.min())
````

### 9. Find the maximum and minimum index values of the output of 7.

````python
# Find arg max
print (output.argmax())

# Find arg min
print (output.argmin())
````

### 10. Make a random tensor with shape `(1, 1, 1, 8)` and then create a new tensor with all the `1` dimensions removed to be left with a tensor of shape `(8)`. Set the seed to `21` when you create it and print out the first tensor and it's shape as well as the second tensor and it's shape.

The output structure should look like:

```
tensor([[[[..., ..., ..., ..., ..., ..., ..., ...]]]]) torch.Size([1, 1, 1, 8])
tensor([..., ..., ..., ..., ..., ..., ..., ...]) torch.Size([8])
```

````python
# Set seed
torch.manual_seed(21)

# Create random tensor
tensor=torch.rand(1,1,1,8)


# Remove single dimensions
tensor_squeezed=tensor.squeeze()

# Print out tensors and their shapes
print(tensor, tensor.shape)
print(tensor_squeezed, tensor_squeezed.shape)
````
